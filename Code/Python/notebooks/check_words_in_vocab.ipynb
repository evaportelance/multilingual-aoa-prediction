```python
import pandas
import pickle
```

```python
word_list_file = "../../../Data/model_datasets/eng/aoa_words.csv"
```

```python
dict_word_file = "../../../Data/model_datasets/eng/encoding_dictionary_vocab_size_5000.pkl"
```

```python
word_lemmas = pandas.read_csv(word_list_file, delimiter="\t")
```

```python
word_lemmas.word_clean
```

```python
with open(dict_word_file, 'rb') as f:
    vocabulary = pickle.load(f)
```

```python
vocabulary
```

```python
in_vocab = []
not_in_vocab = []
for word in word_lemmas.word_clean:
    if word in vocabulary:
        if vocabulary[word] == 0:
            not_in_vocab.append(word)
        else:
            in_vocab.append(word)
    else:
        not_in_vocab.append(word)
```
